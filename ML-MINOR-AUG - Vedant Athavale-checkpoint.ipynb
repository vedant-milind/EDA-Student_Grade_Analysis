```python
sns.catplot(x="gender", y="math_score", hue="test preparation course", kind="boxen", data=df)
sns.catplot(x="gender", y="reading_score", hue="test preparation course", kind="boxen", data=df)
sns.catplot(x="gender", y="writing_score", hue="test preparation course", kind="boxen", data=df)
```

From alll the three graphs above, it is evident that students with completed preparation for exam 
perform better than students with no preparation.

Also, in the maths section, we can see that men score more than women even when they are not prepared well.

In both the other cases of reading and writing, women score a lot better than men even without preparation.

```python
fg= sns.lineplot(x='gender', y='math_score', ci=None, 
             hue='lunch', data=df)
fg.legend(loc='upper right', bbox_to_anchor=(1.4, 1))

```

Here, we can see that in both male and female catgories, students with better meal quality are 
found to be more productive and capable of scoring well in the exams than the students with free
or reduced lunch.

```python

gh= sns.lineplot(x='gender', y='reading_score', ci=None, 
             hue='lunch', data=df)
gh.legend(loc='upper right', bbox_to_anchor=(1.4, 1))

```

Here, the scores for men are found to be less than women overall. Also, students with better lunch, score better.

```python
hj= sns.lineplot(x='gender', y='writing_score', ci=None, 
             hue='lunch', data=df)
hj.legend(loc='upper right', bbox_to_anchor=(1.4, 1))
```

Here, the scores for men are found to be less than women overall. Also, students with better lunch, score better.
